### Counting Simple Matches in Dataset with re()
    
1. Count the number of posts in our data set that match the regex "of Reddit". Assign the count to of_reddit_count.

```python
import re 

of_reddit_count = 0

for i in posts:
    if re.search("of Reddit", i[0]) is not None:
        of_reddit_count += 1
        
print(of_reddit_count)
```

### Using Square Brackets to Match Multiple Characters

1. Use square bracket notation to make the code account for both capitalizations of "Reddit", and count how many posts contain "of Reddit" or "of reddit" in the title.

2. Assign the resulting count to of_reddit_count.

```python
of_reddit_count_old = 0
of_reddit_count = 0
for row in posts:
    if re.search("of Reddit", row[0]) is not None:
        of_reddit_count_old += 1
        
for row in posts:
    if re.search("of [rR]eddit", row[0]) is not None:
        of_reddit_count += 1

print(of_reddit_count)
print(of_reddit_count_old)
```

### Escaping Special Characters

1. Escape the square bracket characters to count the number of posts in our data set that contain the "[Serious]" tag.

2. Assign the count to serious_count.

```python
serious_count = 0

for i in posts:
    if re.search("\[Serious\]", i[0]) is not None:
        serious_count += 1
    
print(serious_count)
```

### Combining Escape Characters And Multiple Matches 
1. Refine the code to count how many posts have either "[Serious]" or "[serious]" in the title.
2. Assign the count to serious_count.

```python
serious_count_old = 0
for row in posts:
    if re.search("\[Serious\]", row[0]) is not None:
        serious_count_old += 1
        
serious_count = 0
for row in posts:
    if re.search("\[[sS]erious\]", row[0]) is not None:
        serious_count += 1
        
print(serious_count)
```

### Adding More Complexity to Your Regular Expression

**Finding Tags posted with "(Serious)" or "(serious)", including the parentheses. Using square bracket notation, and escaping the "[", "]", "(", and ")" characters with the backslash.**

1. Refine the code so that it counts how many posts have the serious tag enclosed in either square brackets or parentheses.
2. Assign the count to serious_count.

```python
serious_count_old = 0
for row in posts:
    if re.search("\[Serious\]", row[0]) is not None:
        serious_count_old += 1
        
serious_count = 0
for row in posts:
    if re.search("[\(\[][sS]erious[\)\]]", row[0]) is not None:
        serious_count += 1
        
print(serious_count)        
```

### Combining Multiple Regular Expressions

1. Use the "^" character to count how many posts include the serious tag at the beginning of the title. Assign this count to serious_start_count.

2. Use the "$" character to count how many posts include the serious tag at the end of the title. Assign this count to serious_end_count.

3. Use the "|" character to count how many posts include the serious tag at either the beginning or end of the title. Assign this count to serious_count_final.

```python
serious_start_count = 0
serious_end_count = 0
serious_count_final = 0

for i in posts:
    if re.search("^[\[\(][Ss]erious[\)\]]", i[0]) is not None:
        serious_start_count += 1
        
for i in posts:
    if re.search("[\[\(][Ss]erious[\)\]]$", i[0]) is not None:
        serious_end_count += 1
        
for i in posts:
    if re.search("^[\[\(][Ss]erious[\)\]]|[\[\(][Ss]erious[\)\]]$", i[0]) is not None:
        serious_count_final += 1
        
print(serious_start_count)
print(serious_end_count)
print(serious_count_final)
```

### Matching Years With Regular Expressions

**"[0-9]" will match any character that falls between 0 and 9 (all of which will be one-digit integers). Similarly, "[a-z]" would match any lowercase letter.**

1. We've loaded a number of strings into the strings variable for you.
2. Loop through strings and use re.search() to determine whether each string contains a year between 1000 and 2999.
3. Store every string that contains a year in year_strings. The .append() function will help here.

```python
strings = ['War of 1812', 'There are 5280 feet to a mile', 'Happy New Year 2016!']

year_strings = []

for i in strings:
    if re.search("[1-2][0-9][0-9][0-9]", i) is not None:
        year_strings.append(i)
        
year_strings
```

### Repeating Characters in Regular Expressions

**We can use curly brackets ("{" and "}") to indicate that a pattern should repeat. To match any four-digit number, for example, we could repeat the pattern "[0-9]" four times by writing "[0-9]{4}"**

1. We've loaded a number of strings into the strings variable for you.
2. Loop through strings and use re.search() to determine whether each string contains a year between 1000 and 2999. Use a regex that takes advantage of curly brackets.
3. Store every string that contains a year in year_strings. The .append() function will help here.

```python
year_strings = []

for i in strings:
    if re.search("[1-2][0-9]{3}", i) is not None:
        year_strings.append(i)
        
year_strings
```
