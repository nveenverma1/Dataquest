### Counting How Many Times a Team Won

### Making a Function that Counts Wins

1. Write a function called nfl_wins() that takes a team name (string value) as the input.
    - The function should return the number of games the team won in the period covered by the data set.
2. Use the function to assign the number of times the Dallas Cowboys ("Dallas Cowboys") won to cowboys_wins.
3. Use the function to assign the number of times the Atlanta Falcons won ("Atlanta Falcons") to falcons_wins.

```python
def nfl_wins(team_str):
    no_of_wins = 0
    for i in nfl:
        if i[2] == team_str:
            no_of_wins += 1
    return no_of_wins

cowboys_wins = nfl_wins("Dallas Cowboys")
print(cowboys_wins)

falcons_wins = nfl_wins("Atlanta Falcons")
print(falcons_wins)
```

3. Create a variable called player_column and set it to the return value of column('player').

```python
# Default display code
class Dataset:
    def __init__(self, data):
        self.header = data[0]
        self.data = data[1:]
        
    # Add your method here.
    def column(self, label):
        idx = -1
        
        for k, v in enumerate(self.header):
            if v == label:
                idx = k

        if idx >= 0:
            result = []
            for i in self.data:
                result.append(i[idx])
            return result
        else:
            return None
    
nfl_dataset = Dataset(nfl_data)

player_column = nfl_dataset.column("player") 
print(player_column)

year_column = nfl_dataset.column("year")
year_column
```
