**The <a href="https://www.bls.gov/">United States Bureau of Labor Statistics (BLS)</a> surveys and calculates the monthly unemployment rate. While the pandas library can read in XLSX files, it relies on an external library for actually parsing the format.<br><br>So, we've instead downloaded the same dataset as a CSV file named unrate.csv, from the website of the <a href="https://www.stlouisfed.org/">Federal Reserve Bank of St. Louis</a>. It has monthly unemployment rate from January 1948 to August 2016.**

### Recap
1. Read unrate.csv into a DataFrame and assign to unrate.
2. Use Pandas.to_datetime to convert the DATE column into a Series of datetime values.
3. Generate a line chart that visualizes the unemployment rates from 1948:
    - x-values should be the first 12 values in the DATE column
    - y-values should be the first 12 values in the VALUE column
4. Use pyplot.xticks() to rotate the x-axis tick labels by 90 degrees.
5. Use pyplot.xlabel() to set the x-axis label to "Month".
6. Use pyplot.ylabel() to set the y-axis label to "Unemployment Rate".
7. Use pyplot.title() to set the plot title to "Monthly Unemployment Trends, 1948".

```python
import pandas as pd
import matplotlib.pyplot as plt

unrate = pd.read_csv('unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])

year_1948 = [unrate['DATE'].head(12), unrate['VALUE'].head(12)]
plt.plot(year_1948[0], year_1948[1])
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
unrate.info()
```

### Matplotlib Classes

1. Use plt.figure() to create a figure and assign to fig.

3. Assign the top Axes object to ax1.
4. Assign the bottom Axes object to ax2.

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
plt.show()
```

### Adding Data
**Create 2 line subplots in a 2 row by 1 column layout:**

1. In the top subplot, plot the data from 1948.
    - For the x-axis, use the first 12 values in the DATE column.
    - For the y-axis, use the first 12 values in the VALUE column.
2. In the bottom subplot, plot the data from 1949.
    - For the x-axis, use the values from index 12 to 24 in the DATE column.
    - For the y-axis, use the values from index 12 to 24 in the VALUE column.<br>

**Use plt.show() to display all the plots.**

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
year_1948 = [unrate['DATE'][:12], unrate['VALUE'][:12]]
ax1.plot(year_1948[0], year_1948[1])

ax2 = fig.add_subplot(2, 1, 2)
year_1949 = [unrate['DATE'][12:24], unrate["VALUE"][12:24]]
ax2.plot(year_1949[0], year_1949[1])

plt.xticks(rotation=90)
plt.show()
```

### Formatting And Spacing 

1. For the plot we generated in the last screen, set the width of the plotting area to 12 inches and the height to 5 inches.

```python
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate[0:12]['DATE'], unrate[0:12]['VALUE'])
ax1.set_title('Monthly Unemployment Rate, 1948')

ax2.plot(unrate[12:24]['DATE'], unrate[12:24]['VALUE'])
ax2.set_title('Monthly Unemployment Rate, 1949')
plt.show()
```

### Comparing Across More Years
1. Set the width of the plotting area to 12 inches and the height to 18 inches.
2. Generate a grid with 5 rows and 1 column and plot data from the individual years. Start with 1948 in the top subplot and end with 1952 in the bottom subplot.
3. Use plt.show() to display the plots.

```python
fig = plt.figure(figsize=(12, 18))

ax1 = fig.add_subplot(5, 1, 1)
year_1948 = [unrate['DATE'][:12], unrate['VALUE'][:12]]
ax1.plot(year_1948[0], year_1948[1])
ax1.set_ylabel("Unemployemnt Rate - 1948")

ax2 = fig.add_subplot(5, 1, 2)
year_1949 = [unrate['DATE'][12:24], unrate["VALUE"][12:24]]
ax2.plot(year_1949[0], year_1949[1])
ax2.set_ylabel("Unemployemnt Rate - 1949")

ax3 = fig.add_subplot(5, 1, 3)
year_1950 = [unrate['DATE'][24:36], unrate['VALUE'][24:36]]
ax3.plot(year_1950[0], year_1950[1])
ax3.set_ylabel("Unemployemnt Rate - 1950")


ax4 = fig.add_subplot(5, 1, 4)
year_1951 = [unrate['DATE'][36:48], unrate['VALUE'][36:48]]
ax4.plot(year_1951[0], year_1951[1])
ax4.set_ylabel("Unemployemnt Rate - 1951")

ax5 = fig.add_subplot(5, 1, 5)
year_1952 = [unrate['DATE'][48:60], unrate['VALUE'][48:60]]
ax5.plot(year_1952[0], year_1952[1])
ax5.set_ylabel("Unemployemnt Rate - 1952")

plt.xlabel("Overall Months")
plt.ylabel("Overall unemployment")
plt.show()
```

### Overlaying Line Charts

1. Set the plotting area to a width of 6 inches and a height of 3 inches.
2. Generate 2 line charts in the base subplot, using the MONTH column for the x-axis instead of the DATE column:
    - One line chart using data from 1948, with the line color set to "red".
    - One line chart using data from 1949, with the line color set to "blue".
3. Use plt.show() to display the plots.

```python
unrate['MONTH'] = unrate['DATE'].dt.month
# print(unrate['MONTH'])

fig = plt.figure(figsize=(6, 3))
year_1948 = [unrate['MONTH'][:12], unrate['VALUE'][:12]]
year_1949 = [unrate['MONTH'][12:24], unrate['VALUE'][12:24]]

plt.plot(year_1948[0], year_1948[1], c='red')
plt.plot(year_1949[0], year_1949[1], c='blue')

plt.show()
```

### Adding More Lines

1. Set the plotting area to a width of 10 inches and a height of 6 inches.
2. Generate the following plots in the base subplot:
    - 1948: set the line color to "red"
    - 1949: set the line color to "blue"
    - 1950: set the line color to "green"
    - 1951: set the line color to "orange"
    - 1952: set the line color to "black"
3. Use plt.show() to display the plots.

```python
fig = plt.figure(figsize=(10, 6))

year_1948 = [unrate['MONTH'][:12], unrate['VALUE'][:12]]
plt.plot(year_1948[0], year_1948[1], c='red')

year_1949 = [unrate['MONTH'][12:24], unrate["VALUE"][12:24]]
plt.plot(year_1949[0], year_1949[1], c='blue')

year_1950 = [unrate['MONTH'][24:36], unrate['VALUE'][24:36]]
plt.plot(year_1950[0], year_1950[1], c='green')

year_1951 = [unrate['MONTH'][36:48], unrate['VALUE'][36:48]]
plt.plot(year_1951[0], year_1951[1], c='orange')

year_1952 = [unrate['MONTH'][48:60], unrate['VALUE'][48:60]]
plt.plot(year_1952[0], year_1952[1], c='black')

plt.show()
```

### Adding a Legend

1. Modify the code from the last screen that overlaid 5 plots to include a legend. Use the year value for each line chart as the label.
    - E.g. the plot of 1948 data that uses "red" for the line color should be labeled "1948" in the legend.
2. Place the legend in the "upper left" corner of the plot.
3. Display the plot using plt.show().

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
labels = [1948, 1949, 1950, 1951, 1952]
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=labels[i])
    plt.legend(loc="upper left")
plt.show()
```

### Final Tweaks

1. Modify the code from the last screen:
    - Set the title to "Monthly Unemployment Trends, 1948-1952".
    - Set the x-axis label to "Month, Integer".
    - Set the y-axis label to "Unemployment Rate, Percent".

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
plt.legend(loc='upper left')
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.show()
```
