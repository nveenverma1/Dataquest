# *Bar Plots And Scatter Plots*

**<a href="https://www.dataquest.io/m/144/bar-plots-and-scatter-plots/2/www.fivethirtyeight.com">FiveThirtyEight</a> has compiled data for 147 films from 2015 that have substantive reviews from both critics and consumers. Every time Hollywood releases a movie, critics from Metacritic, Fandango, Rotten Tomatoes, and IMDB.<br>We will investigate this dataset to see if there is any bias to Fandango's ratings. In addition to aggregating ratings for films, Fandango is unique in that it also sells movie tickets, and so it has a direct commercial interest in showing higher ratings.<br>We'll be working with the fandango_scores.csv file, which can be downloaded from the <a href="https://github.com/fivethirtyeight/data/tree/master/fandango">FiveThirtEight Github repo.</a>**

### Data Dictionary
> <font color=blue>***FILM***</font> - film name<br>
> <font color=blue>***RT_user_norm***</font> - average user rating from Rotten Tomatoes, normalized to a 1 to 5 point scale<br>
> <font color=blue>***Metacritic_user_nom***</font> - average user rating from Metacritc, normalized to a 1 to 5 point scale<br>
> <font color=blue>***IMDB_norm***</font> - average user rating from IMDB, normalized to a 1 to 5 point scale<br>
> <font color=blue>***Fandango_Ratingvalue***</font> - average user rating from Fandango, normalized to a 1 to 5 point scale<br>
> <font color=blue>***Fandango_Stars***</font> - the rating displayed on the Fandango website (rounded to nearest star, 1 to 5 point scale)


### Introduction to the Data
1. Read **fandango_scores.csv** into a Dataframe named reviews.
2. Select the following columns and assign the resulting Dataframe to norm_reviews:
    - [FILM, RT_user_norm, Metacritic_user_nom (note the misspelling of norm),<br>IMDB_norm, Fandango_Ratingvalue, Fandango_Stars]
3. Display the first row in norm_reviews

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

reviews = pd.read_csv("fandango_scores.csv")
norm_reviews = reviews.loc[:,['FILM', 'RT_user_norm', 
                          'Metacritic_user_nom', 
                          'IMDB_norm', 'Fandango_Ratingvalue', 
                          'Fandango_Stars']]
print(type(norm_reviews))
norm_reviews[:5]
```

### Creating Bars
1. Create a single subplot and assign the returned Figure object to fig and the returned Axes object to ax.
2. Generate a bar plot with:
    - left set to bar_positions
    - height set to bar_heights
    - width set to 0.5
3. Use plt.show() to display the bar plot.

```python
from numpy import arange

num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_heights = norm_reviews[num_cols].iloc[0].values
bar_positions = arange(5) + 0.75

fig, ax = plt.subplots()

ax.bar(bar_positions, bar_heights, 0.5, align='edge')

plt.show()
```

### Aligning Axis Ticks and Labels

1. Create a single subplot and assign the returned Figure object to fig and the returned Axes object to ax.
2. Generate a bar plot with:
    - left set to bar_positions
    - height set to bar_heights
    - width set to 0.5
3. Set the x-axis tick positions to tick_positions.
4. Set the x-axis tick labels to num_cols and rotate by 90 degrees.
5. Set the x-axis label to ***"Rating Source"***.
6. Set the y-axis label to ***"Average Rating"***.
7. Set the plot title to ***"Average User Rating For Avengers: Age of Ultron (2015)"***.
8. Use plt.show() to display the bar plot.

```python
fig, ax = plt.subplots()

bar_position = arange(5) + 0.75
bar_height = norm_reviews[num_cols].iloc[0].values

ax.bar(bar_position, bar_height, 0.5, align='edge')

tick_positions = range(1, 6)
ax.set_xticks(range(1, 6))

ax.set_xticklabels(num_cols, rotation=90)

plt.xlabel("Rating Source")
plt.ylabel("Average Rating")
plt.title("Average User Rating For Avengers: Age of Ultron (2015)")

plt.show()
```

### Frequency Distribution
1. Use the *value_counts()* method to return the frequency counts for the ***Fandango_Ratingvalue*** column. Sort the resulting Series object by the index and assign to ***fandango_distribution***.
2. Use the *value_counts()* method to return frequency counts the ***IMDB_norm*** column. Sort the resulting Series object by the index and assign to ***imdb_distribution***.
3. Use the *print()* function to display ***fandango_distribution*** and ***imdb_distribution***.

```python
fandango_distribution = norm_reviews['Fandango_Ratingvalue'].value_counts().sort_index()
print(fandango_distribution, end="\n\n")

imdb_distribution = norm_reviews['IMDB_norm'].value_counts().sort_index()
print(imdb_distribution)
```
