### Math Behind Slope

```python
def slope(x1,x2,y1,y2):
    ## Finish the function here
    return (y2-y1)/(x2-x1)

slope_one = slope(0,4,1,13)
slope_two = slope(5,-1,16,-2)

print(slope_one)
print(slope_two)
```

### Secant Lines

```python
import seaborn
seaborn.set(style='darkgrid')

def draw_secant(x_values):
    x = np.linspace(-20, 30, 100)
    y = -(x**2) + (x*3) - 1
    plt.plot(x, y)
    
    # Drawing Secant Line
    x1 = x_values[0]
    x2 = x_values[1]
    
    y1 = -(x1**2) + (x1*3) - 1
    y2 = -(x2**2) + (x2*3) - 1
    
    m = (y2-y1)/(x2-x1)
    b = y1 - m*x1
    
    y = m*x + b
    plt.plot(x, y, color='green')
    plt.show()
    
draw_secant([3, 5])
draw_secant([3, 10])
draw_secant([3, 15])
```

### Properties of Limits I

$\lim_{x \to a} [f(x) + g(x)] = \lim_{x \to a} f(x) + \lim_{x \to a} g(x)$<br>
$\lim_{x \to a} [f(x) - g(x)] = \lim_{x \to a} f(x) - \lim_{x \to a} g(x)$

E.g.<br>

$$\lim_{x \to 1} [3x^2 + 3x -3]$$<br> is same as <br>
$$\lim_{x \to 1} [3x^2] + \lim_{x \to 1} [3x] - \lim_{x \to 1} [3]$$<br> which is equal to 3 + 3 -3 = 3


```python
x,y = sympy.symbols('x y')

approaches = 1
input_v = x
func = 3*(x**2) + 3*x - 3

limit_two = sympy.limit(func, input_v, approaches)
limit_two
```

### Properties of Limits II

$\lim_{x \to a} [cf(x)] = c \lim_{x \to a}f(x)$

E.g.<br>

$$\lim_{x \to 1} [3x^2 + 3x -3]$$<br> is same as <br>
$$3\lim_{x \to 1} [x^2] + 3\lim_{x \to 1} [x] - 3\lim_{x \to 1} [1]$$<br> which is equal to 3 + 3 -3 = 3

```python
x,y = sympy.symbols('x y')

approaches = -1
input_v = x
func = x**3 + 2*(x**2) - 10*x

limit_three = sympy.limit(func, input_v, approaches)
limit_three
```

### Undefined Limit to Defined Limit

$$\lim_{x_2 \to 3} \frac{f(x_2)+1}{x_2-3} = \lim_{x_2 \to 3} \frac{-x_2^2+3x_2-1+1}{x_2-3}$$

using $f(x) = -x^2 + 3x -1$ simplyfying to

$$\lim_{x_2 \to 3} \frac{-x_2^2 + 3x_2}{x_2-3}$$

still the denominator becomes zero, and equation becomes undefined. So lets simplify more

$$\lim_{x_2 \to 3} \frac{-x_2(x_2 - 3)}{x_2-3}$$

or 

$$\lim_{x_2 \to 3} -x_2$$

```python
x2,y = sympy.symbols('x y')

approaches = 3
input_v = x2
func = -x2

limit_four = sympy.limit(func, input_v, approaches)
limit_four
```

# *Finding Extreme Points*

### Differentiation

A **Derivative** is the slope of a function at any given point.

$$\lim_{h \to 0} \frac{f(x+h)-f(x)}{h}$$

Instead of selecting a specific $x$ value, we'll leave $x$ where it is and expand the equation to solve for the derivative.<br> Since $f(x) = -(x^2) + 3x -1$, this works out to:

$$\lim_{h \to 0} \frac{(-(x+h)^2 + 3(x+h) -1)-(-(x^2) + 3x -1)}{h}$$

Expanding it, we get:

$$\lim_{h \to 0} \frac{-x^2 - 2xh - h^2 + 3x+ 3h -1 + x^2 - 3x +1}{h}$$

After cancelling terms, we get:

$$\lim_{h \to 0} \frac{-2xh - h^2 + 3h}{h}$$

Simplifying it further:

$$\lim_{h \to 0} \frac{h(-2x - h + 3)}{h}$$

And then:

$$\lim_{h \to 0} -2x - h + 3$$

Further taking the limit, slope of the function `f(x)`, becomes `-2x + 3`

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Below is the graph of f(x)
# The function curves from positive to negative at x = 1.5
x = np.linspace(0, 3, 50)
y = -(x**2) + (x*3) - 1
plt.plot(x, y)
plt.grid()
plt.show()

# Below is the graph for slope of f(x)
# The slope transitions from positive to negative at x = 1.5
x = np.linspace(0, 3, 50)
y = -2*x + 3
plt.plot(x, y)
plt.grid()
plt.show()
```

### Power Rule

$$f(x) = x^r$$

$$f'(x) = rx^{r-1}$$

If our original function is $f(x) = x^2$, $r$ would be 2. Its derivative would be $f'(x) = 2x^{2-1}$ or $f'(x) = 2x$ for short. The function for our original curve is a similar value ($-x^2$) added to $3x-1$:

$f(x) = -(x)^2 + 3x - 1$

When we calculate the derivative of a function, we're really computing and solving a limit. We can use the properties of limits to break down the function into 3 different derivatives. We can then apply the power rule to each component:

$f'(x) = -2x + 3$

### Linearity Of Differentiation

Linearity of differentiation consists of 2 rules. First is the sum rule:

$\frac{d}{dx}[f(x) + g(x)] = \frac{d}{dx}[f(x)] + \frac{d}{dx}[g(x)]$

Second is the constant factor rule, which lets us pull out constants from the derivative:

$\frac{d}{dx}[cf(x)] = c\frac{d}{dx}[f(x)]$

By combining both of these rules, we can find the derivative for a wide range of functions. For example:

$\frac{d}{dx}[-x^3 + 3x^2] = -\frac{d}{dx}[x^3] + 3\frac{d}{dx}[x^2]  = -3x^2 + 6x$

```python
x3 = 1
fx3 = (x3**5) - x3
slope_three = 5*(x3**4) - 1

x4 = 2
fx4 = (x4**3) - (x4**2)
slope_four = 3*(x4**2) - 2*x4

print(slope_three)
print(slope_four)
```
