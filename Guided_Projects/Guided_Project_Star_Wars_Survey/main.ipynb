```python
import pandas as pd
star_wars = pd.read_csv('StarWars.csv', encoding='ISO-8859-1')
star_wars = star_wars[star_wars['RespondentID'].notnull()]
star_wars
```

### Cleaning and Mapping Yes/No Columns

1. Convert the Have you seen any of the 6 films in the Star Wars franchise? column to the Boolean type.
2. Convert the Do you consider yourself to be a fan of the Star Wars film franchise? column to the Boolean type.
3. When you're finished, both columns should only contain the values True, False, and NaN.

```python
map_dict = {'Yes': True, 'No': False}

cols = star_wars.columns[1:3]

for i in cols:
    star_wars[i] = star_wars[i].map(map_dict)

star_wars
```

### Cleaning and Mapping Checkbox Columns

1. Convert each column above so that it only contains the values True and False.
    - You can select the column names more quickly by entering star_wars.columns[3:9], rather than typing them out.
    - Be very careful with spacing when constructing your mapping dictionary! In the cells, Star Wars: Episode I The Phantom Menace has two spaces between the end of Episode I and the start of The Phantom, but this is not the case in Star Wars: Episode VI Return of the Jedi. Check the values in the cells carefully to make sure you use the appropriate spacing.
2. Rename each of the columns above so the names are more intuitive. We recommend using seen_1 to indicate whether the respondent saw Star Wars: Episode I The Phantom Menace, seen_2 for Star Wars: Episode II Attack of the Clones, and so on.
3. When you're finished, the columns should have intuitive names, along with True and False values that indicate whether the respondent saw each of the six Star Wars movies.

```python
import numpy as np

col_to_rename = star_wars.columns[3:9]


naming_dict = {'Star Wars: Episode I  The Phantom Menace': True,
               'Star Wars: Episode II  Attack of the Clones': True,
               'Star Wars: Episode III  Revenge of the Sith': True,
               'Star Wars: Episode IV  A New Hope': True,
               'Star Wars: Episode V The Empire Strikes Back': True,
               'Star Wars: Episode VI Return of the Jedi': True,
               np.NaN: False}

for i in col_to_rename:
    star_wars[i] = star_wars[i].map(naming_dict)
    
star_wars = star_wars.rename(columns={col_to_rename[0] : 'seen_1',
                                     col_to_rename[1] : 'seen_2',
                                     col_to_rename[2] : 'seen_3',
                                     col_to_rename[3] : 'seen_4',
                                     col_to_rename[4] : 'seen_5',
                                     col_to_rename[5] : 'seen_6'})
star_wars
```

### Finding The Highest-Ranked Movie

1. Use the pandas.DataFrame.mean() method to compute the mean of each of the ranking columns from the last screen.

```python
import matplotlib.pyplot as plt
%matplotlib inline

print(star_wars[star_wars.columns[9:15]].mean())

index = np.arange(6) 
plt.bar(index, star_wars[star_wars.columns[9:15]].mean())
plt.show()
```

### Explore the Data by Binary Segments

1. Split the data into two groups based on one of the binary columns above.
2. Redo the two previous analyses (find the most viewed movie and the highest-ranked movie) separately for each group, and then compare the results.
3. If you see any interesting patterns, write about them in a Markdown cell.

```python
males = star_wars[star_wars['Gender']=='Male']
females = star_wars[star_wars['Gender']=='Female']

fig = plt.figure(figsize=(12, 8))

ax1 = fig.add_subplot(2, 2, 1)
ax1.bar(index, males[males.columns[9:15]].mean())
ax1.set_title('Male Rankings')

ax2 = fig.add_subplot(2, 2, 2)
ax2.bar(index, females[females.columns[9:15]].mean())
ax2.set_title('Female Rankings')

ax3 = fig.add_subplot(2, 2, 3)
ax3.bar(index, males[males.columns[3:9]].sum())
ax3.set_title('Male Views')

ax4 = fig.add_subplot(2, 2, 4)
ax4.bar(index, females[females.columns[3:9]].sum())
ax4.set_title('Female Views')
plt.show()
```

**Here, we have an interesting observation that more males watched the newer movies, but they liked it lesser then their female counterparts.**
